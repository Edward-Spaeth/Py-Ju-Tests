```python
from graphviz import Digraph
# Create Digraph object
dot = Digraph()


Nodes = ['A', 'B', 'C', 'D', 'A', 'E']

Edges = [('A', 'C'), ('B', 'A'), ('B', 'C'), ('C', 'B'), ('A', 'C'), ('F', 'G')] # Fügt fälschlicherweise Knoten F und G hinzu (Achtung!), obwohl nicht in Nodes definiert

for Node in Nodes:
    dot.node (Node)

for Start,Dest in Edges:
    dot.edge (Start,Dest)
    
dot
```
